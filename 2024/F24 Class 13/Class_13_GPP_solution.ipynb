Create a list of numbers. Get the sum of the numbers using the built-in **sum** function, and also by creating your own running sum in a loop.

````python
# 2
numlist = [ 3, 11, 5, 4, -9]
print(sum(numlist))
print()
running_sum = 0
for num in numlist:
    running_sum = running_sum + num
print(running_sum)
````

Create a list of numbers. Find the minimum using the built-in **min** function, and also by using a loop.

````python
# 3
numlist = [ 3, 11, 5, 4, -9]
print(min(numlist))
print()
minimum = numlist[0]
for num in numlist:
    if num < minimum:
        minimum = num
print(minimum)
````

Create a long string. Find how many spaces there are in the string using the **count** method, and also by looping to test every character.

````python
# 4
sentence = "The quick brown fox jumped over the lazy dog."
print(sentence.count(' '))
print()
count = 0
for char in sentence:
    if char == ' ':
        count = count + 1
print(count)
````

Create a list. Reverse the items in the list, and store in a new variable. Do this 2 ways: using indexing, and using a loop.

````python
# 5
mylist = [4, 33, 11, 5, 2]
print(mylist)
new_list = mylist[::-1]
print(new_list)
new_list = []
for i in range(len(mylist)-1, -1, -1):
    new_list.append(mylist[i])
print(new_list)
````

Create a list of strings. From this, create a list of the lengths of the strings. Do this 2 ways: using a list comprehension and using a **for** loop.

*Comprehensions* are useful as ways to simplify code. However, we have to be careful about writing WORN (**Write Once, Read Never**) code, as the *comprehension* syntax is unique to Python and so non-Python speakers will not understand it.

So be judicious in your use of comprehensions, and don't use a single-line comprehension just to avoid writing a for loop across two lines.

For instance, consider how much more language-specific knowledge it takes to understand:

```
numList = [1,2,3,4,5]
[i ** 2 for i in numList]
```

vs.

```
numList = [1,2,3,4,5]
newList = numList
for i in range(len(numList)):
  numList[i] = numList[i] ** 2
```

The second is wordier, and that helps non-native python speakers, as the `for` loop construction is universal in all languages. Someone familiar with programming in general -- but not python in specific-- would be able to trivially understand the `for` loop code. Whereas the comprehension would be, well, *incomprehensible*.

With this in mind, write the following code as nested for loops and compare how readable the two variants are.

```
matrix = [[[k for k in range(3)] for j in range(3)] for i in range(3)]
print(matrix)
```

````python
#10
newMatrix = []
for i in range(3):
  newJMatrix = [];
  for j in range(3):
    newJMatrix.append(list(range(3)))
  newMatrix.append(newJMatrix)

print(newMatrix)
````

Generate a set of vowels from a given string using a set comprehension. Hint, this will require using an `if` as part of the comprehension
