Today we will investigate the NumPy module. NumPy is short for Numerical Python, and it is based on arrays.

Start by executing the following, to import numpy as np and to import the array function.

```python
# 1
import numpy as np
from numpy import array
```

Create a one-dimensional array variable using the array function, in which 5 integers are stored. Also create a 1D array variable that stores 6 floats. Print both arrays and their types.

```python
# 2
int_array = array([11, 2, 33, 9, 4])
float_array = array([1.1, 2.2, 3.3, 4.8, 5.4, 6.2])
print(int_array, type(int_array))
print(float_array, type(float_array))
```

Use the dtype method to determine the type of both array variables. Note that no arguments are passed to dtype, and the parentheses are not needed.

```python
# 3
print(int_array.dtype)
print(float_array.dtype)
```

For both variables, use the ndim method to determine the number of dimensions, the shape method to determine the length of the axes, and the size method to determine the number of elements.

```python
# 4
print(int_array.ndim)
print(float_array.ndim)
print(int_array.shape)
print(float_array.shape)
print(int_array.size)
print(float_array.size)
```

OK, time for two dimensional (2D) arrays!  First, create a nested list variable that stores integers: two inner lists, each of which stores 3 integers. Then, create a 2 by 3 array variable by passing the list to the array function.  Print both variables so you can see the difference.

```python
# 5

nested_list = [[1, 2, 3], [4, 5, 6]]
two_d_array = array(nested_list)
print(nested_list)
print(two_d_array)
```

Use the ndim, shape, and size methods on your 2D array variable.

```python
# 6

print(two_d_array.ndim)
print(two_d_array.shape)
print(two_d_array.size)
```

Create a 1D array that stores [2  5  8  11]. Do this two ways: using **arange** and **linspace**.  Make sure that the elements in the array are integers.

```python
# 7

my1d = np.arange(2, 12, 3)
print(my1d)
my1d = np.linspace(2, 11, 4, dtype = np.int16)
print(my1d)
```

Create a 2 by 4 array using **arange** for the numbers on the first row, and **linspace** for the numbers on the second row. Print it.

```python
# 8

my2d = array([np.arange(2, 9, 2), np.linspace(1, 8, 4, dtype = np.int16)])
print(my2d)
```

Create and print a 3 by 5 array of all zeros (integers).

Create and print a 4 by 2 array of all ones (integers).

Execute the following to create a random number generator, and then use that to create arrays of random floats and integers.

```python
# 12
rng = np.random.default_rng()
randfl = rng.random(4)
print(randfl)
randint = rng.integers(0, 100, 5)
print(randint)
```

Create a 2 by 4 array of random integers, in the range from 0 to 50. Hint: instead of a length of 5 as above, the third argument will be a tuple that contains the number of rows and columns.

```python
# 13
randints = rng.integers(0, 50, (2,4))
print(randints)
```

Create a 1D array and practice indexing into it and slicing, using negative and positive indices.

```python
# 14
my1d = array([33, 2, 11, 5, 8, 4])
print(my1d[0])
print(my1d[1])
print(my1d[-1])
print(my1d[::2])
print(my1d[1::2])
print(my1d[::-1])
```

Create a 2D array and practice indexing into it and slicing.

```python
# 15
my2d = array([[1, 22, 32], [14, 5, 9]])
print(my2d)
print(my2d[0, 0])
print(my2d[0, 1])
print(my2d[0, -1])
print(my2d[1, :])
print(my2d[:, 0])
print(my2d[:, :2])
```

Create a 2D array and replace the first row with other numbers.

```python
# 16
my2d = array([[1, 22, 32], [14, 5, 9]])
print(my2d)
my2d[0, :] = [1, 2, 3]
print(my2d)
```

Using an array in a Boolean expression results in a logical array. Create a 1D array variable my1d and try the expression my1d < 10. (Store it in a variable and print it.)

```python
# 17
my1d = array([33, 2, 11, 5, 8, 4])
mylog = my1d < 10
print(mylog)
```

Now, use your logical array to index into your original 1D array variable.

```python
# 18
print(my1d[mylog])
```

Create a 2 by 4 array. Use the reshape method to reshape it into as many other arrays as you can.

```python
# 19
my2d = array([[1, 22, 32, 1], [14, 5, 9, 2]])
print(my2d)
print(my2d.reshape(4, 2))
print(my2d.reshape(1,8))
print(my2d.reshape(8,1))
```

Create a 2D array and transpose it. Print both.

```python
# 20
my2d = array([[1, 22, 32, 1], [14, 5, 9, 2]])
print(my2d)
print(my2d.T)
```
