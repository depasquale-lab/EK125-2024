4. Remove the `'year_released'` key using the `pop()` method.

You are given a dictionary representing the stock of items in a grocery store:

```python
store_stock = {
    'apples': 50,
    'bananas': 30,
    'milk': 20,
    'bread': 15
}
```
1. Use a loop to print each item and its stock in the format: `"Item: <item>, Stock: <stock>"`.
2. Use the `.keys()` method to print a list of all items in the store.
3. Use the `.values()` method to calculate and print the total stock of all items.
4. Use the `.items()` method to print each item and its stock as tuples.

````python
#7
# Step 1: Define the dictionary
store_stock = {
    'apples': 50,
    'bananas': 30,
    'milk': 20,
    'bread': 15
}

# Step 2: Loop to print each item and its stock
print("Stock details:")
for item, stock in store_stock.items():
    print(f"Item: {item}, Stock: {stock}")

# Step 3: Use .keys() to print all items
items = list(store_stock.keys())
print("\nItems in store:", items)

# Step 4: Use .values() to calculate total stock
total_stock = sum(store_stock.values())
print("\nTotal stock:", total_stock)

# Step 5: Use .items() to print item-stock tuples
print("\nItem-stock tuples:")
for item_tuple in store_stock.items():
    print(item_tuple)
````

You are managing data about students and their test scores. The data is stored in a dictionary where each key is a student's name, and the value is another dictionary containing the student's scores for Math, Science, and English.

Here is the data:

```python
students_scores = {
    'Alice': {'Math': 85, 'Science': 90, 'English': 88},
    'Bob': {'Math': 78, 'Science': 82, 'English': 84},
    'Charlie': {'Math': 92, 'Science': 88, 'English': 91},
    'Dana': {'Math': 89, 'Science': 94, 'English': 87}
}
```

1. Print the scores for `'Alice'`.
2. Retrieve and print the Science score for `'Charlie'`.
3. Add a new student, `'Eve'`, with the following scores:
- Math: 88
- Science: 86
- English: 89
4. Calculate and print the average Math score for all students.
5. Use a loop to print each studentâ€™s name and their average score in the format: `"Student: <name>, Average Score: <average>"`.

````python
#8
# Step 1: Define the dictionary
students_scores = {
    'Alice': {'Math': 85, 'Science': 90, 'English': 88},
    'Bob': {'Math': 78, 'Science': 82, 'English': 84},
    'Charlie': {'Math': 92, 'Science': 88, 'English': 91},
    'Dana': {'Math': 89, 'Science': 94, 'English': 87}
}

# Step 2: Print the scores for 'Alice'
print("Alice's Scores:", students_scores['Alice'])

# Step 3: Retrieve and print the Science score for 'Charlie'
charlie_science = students_scores['Charlie']['Science']
print("\nCharlie's Science Score:", charlie_science)

# Step 4: Add a new student, 'Eve'
students_scores['Eve'] = {'Math': 88, 'Science': 86, 'English': 89}

# Step 5: Calculate the average Math score
math_scores = [scores['Math'] for scores in students_scores.values()]
average_math_score = sum(math_scores) / len(math_scores)
print("\nAverage Math Score:", average_math_score)

# Step 6: Loop to print each student's name and average score
print("\nStudent Average Scores:")
for student, scores in students_scores.items():
    average_score = sum(scores.values()) / len(scores)
    print(f"Student: {student}, Average Score: {average_score:.2f}")
````

You are tasked with creating a DataFrame to store information about students and their scores in Math, Science, and English.

1. Create a DataFrame named `students_scores` using the following data:

- Science: `[88, 76, 85, 95, 89]`
- English: `[92, 85, 89, 94, 90]`
2. Print the entire DataFrame.

4. Retrieve and print the scores for `'Dana'` across all subjects.
5. Add a new column named `'Total'` that contains the total score for each student.

````python
#12
# Step 2: Create the DataFrame
students_scores = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Dana', 'Eve'],
    'Math': [85, 90, 78, 92, 88],
    'Science': [88, 76, 85, 95, 89],
    'English': [92, 85, 89, 94, 90]
})

# Step 3: Print the entire DataFrame
print("Students Scores DataFrame:")
print(students_scores)

# Step 4: Print the Math column
print("\nMath Scores:")
print(students_scores['Math'])

# Step 5: Retrieve and print scores for 'Dana'
dana_scores = students_scores.loc[students_scores['Name'] == 'Dana']
print("\nDana's Scores:")
print(dana_scores)

# Step 6: Add a Total column
students_scores['Total'] = students_scores['Math'] + students_scores['Science'] + students_scores['English']

# Step 7: Print the updated DataFrame
print("\nUpdated DataFrame with Total Scores:")
print(students_scores)
````
