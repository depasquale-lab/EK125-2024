Write a function called `countWords` which takes `filename` as an argument, and which counts the number of words in the file. Run this on `sample.txt`.

(HINT: the total word count of `sample.txt` is `16`)
(HINTHINT: Last week we reviewed the method `split()`. We suggest you use it here!)

````python
# 4
def countWords(filename):
    with open(filename, 'r') as file:
        content = file.read()
        words = content.split()
        return len(words)

wordCount = countWords('sample.txt')
print(f'Total words: {wordCount}')
````

Write a Python function `processFile` which takes as arguments `inputFilename` and `outputFilename`, and which creates a new file named `output.txt`, writing the analyzed data (word count, line count, character count) into it.

NOTE: You cannot pass multiple arguments to `write()` like you can to `print()`. You'll either have to call `write()` multiple times or pass it a formatted string (You **do** remember how to use`'f`, right???)

NOTENOTE: Do this with only a single "atomic" write. It's very inefficient to continually open and close files, and generally you want to avoid doing this midstream unless there are specific reasons (such as to "flush" data in case data loss is a risk.)

When you're done, `output.txt` should look like

```
Total words: 16
Total lines: 3
Total characters: 101
```

You can see how `output.txt` turned out by looking at it in the file browser on Colab's left side.

````python
# 5
def processFile(inputFilename, outputFilename):
    with open(inputFilename, 'r') as file:
        content = file.read()
        words = content.split()
        lines = content.split('\n')
        characters = len(content)

    with open(outputFilename, 'w') as file:
      file.write(f'Total words: {len(words)}\n')
      file.write(f'Total lines: {len(lines)}\n')
      file.write(f'Total characters: {characters}')

processFile('sample.txt', 'output.txt')
````
