Write a Python function to count the number of lines, words, and characters in `example.txt`.

````python
# 5
def count_file_stats(filename):
    f = open(filename, 'r')
    lines = f.readlines()
    f.close()
    num_lines = len(lines)
    num_words = sum(len(line.split()) for line in lines)
    num_chars = sum(len(line) for line in lines)
    return num_lines, num_words, num_chars

lines, words, chars = count_file_stats('example.txt')
print('Lines: ', lines, ', Words: ', words, ', Characters: ', chars)
````

Create a text file named `log.txt` with the following content:

```
INFO: User logged in
ERROR: Failed to load resource
INFO: User performed an action
WARNING: Low disk space
INFO: User logged out
ERROR: Failed to save file
WARNING: High memory usage
INFO: User logged in
```

Write a Python script to read the log.txt file and count the occurrences of each log level (INFO, ERROR, WARNING).
Use a loop to iterate through each line of the file.
Use enumeration to keep track of the line numbers.

````python
#10

def analyze_log(filename):
    # Initialize counters for each log level
    info_count = 0
    error_count = 0
    warning_count = 0

    # Open the file in read mode
    with open(filename, 'r') as f:
        # Loop through each line with enumeration
        for line_number, line in enumerate(f, start=1):
            # Check the log level and update the corresponding counter
            if line.startswith("INFO"):
                info_count += 1
            elif line.startswith("ERROR"):
                error_count += 1
            elif line.startswith("WARNING"):
                warning_count += 1

            # Print the line number and the line content
            print(f"Line {line_number}: {line.strip()}")

    # Print the summary report
    print('Summary Report:')
    print('INFO: ', info_count)
    print('ERROR: ',  error_count)
    print('WARNING: ', warning_count)

# Analyze the log file
analyze_log('log.txt')
````
