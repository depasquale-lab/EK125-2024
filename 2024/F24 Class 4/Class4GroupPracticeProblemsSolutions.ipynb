We can also use formatted strings to substitute the value of an expression into a string.  To do this we'll add an 'f' or 'F' before the string and use curly braces around the name of the expression to insert in our print statement.  The part that makes it a formatted string, are format specifiers.  These we'll put after a a colon and the expression name.  For example:   

```python
unexpected1 =input('Please enter an unexpected thing: ')
unexpected2 =input('And another unexpected thing: ')
unexpected3 =input('And a third, please: ')
print('Nobody expects the', unexpected1, '!')
print(f'Nobody expects the {unexpected2}!')
print(f'Nobody expects the{unexpected3:25s}!')
```

Notice in the statements above that if we don't use formatted strings at all, a space is added between strings when printing.

Formatted strings let us control the spacing by controlling the width of the field that the value of the expression is placed in.

Strings, like those above (with a specifier of 's') are left justified in that field.

Integers (with a specifier of 'd'), however, are right justified.

```python
mystr=input("Please enter a three digit number")
myint=int(mystr)
print("The integer entered is", myint, ".")
print(f"The integer entered is {myint}.")
print(f"The integer entered is {myint:8d}.")
```

Floats are a little bit more complicated.  They have a spcifier of 'f'.  Instead of specifying just a field width, floats also require the number of decimal places, so '6.3f' would have a field width of 6 including 3 decimal places and the decimal point.

Two additional specifiers exist for floats.  'g' specifies a general format and 'e' specifies scientific notation

```python
mystr=input("Please enter a number with three digits before and three digits following a decimal point")
myfloat=float(mystr)
print(f"The float entered is {myfloat:8.2f}")
print(f"The float entered is {myfloat:.2f}")
print(f"The float entered is {myfloat:f}")
print(f"The float entered is {myfloat:g}")
print(f"The float entered is {myfloat:e}")
```

Any formatted string can also be stored as a variable.  Try storing the string you entered above as a dollar amount.

```python
dollar=print(f'${myfloat:.2f}')
dollar
```

Let's say we're trying to automate the process for estimating how many cans of paint our customers need to buy when painting a room.  Let's work our way through how this might work.  

We know that one can of paint covers 400 square feet of wall surface.  So we'll need to know the total surface area of the walls to paint.  Assume a rectangular room with no windows or doors.

You'll need to ask the customer for three things (remember units!).

Start by writing your algorithm in comments.

```python
# This script calculates the number of gallons of paint required to paint a room
# and prints out a sentence with this information to the user

# Ask the user for: length of floor, width of floor, height of wall in feet

# Calculate the surface area of the two pairs of walls

# Find the total surface area by summing these.

# Find the number of cans of paint necessary by dividing by 400 and rounding up to the nearest integer.

# Print this information to the user in a complete sentence.
```

Now that you have your steps laid out in comments, it should be relatively simple to carry out these steps in code.  Give that a try.  Copy your comments from the cell above and fill in between them.  Do any prompting for input, calculating, formatting output and printing necessary.

```python
# This script calculates the number of gallons of paint required to paint a room
# and prints out a sentence with this information to the user
import math

# Ask the user for: length of floor, width of floor, height of wall in feet
length=input('Please enter the length of the floor (in ft): ')
width=input('Please enter the width of the floor (in ft)')
height=input('Please enter the height of the walls (in ft)')

# Calculate the surface area of the two pairs of walls
wall1=2*float(length)*float(height)
wall2=2*float(width)*float(height)

# Find the total surface area by summing these.
total_area=wall1+wall2

# Find the number of cans of paint necessary by dividing by 400 and rounding up to the nearest integer.
cans=math.ceil(total_area/400)

# Print this information to the user in a complete sentence.
print(f'The number of cans of paint required for your project is {cans}.')
```
