A CNC milling machine has a file of instructions that basically tell it to move from one point in 3D space to the next, then the next, etc. This consists of a file in which each row has the x,y, and z coordinates in a line in the file with the format 'x12.34y3.5z4.21'. The format is the letter 'x' followed by the x coordinate, the letter 'y', the y coordinate, the letter 'z' and the z coordinate. The number of decimal places in the coordinates can vary. We will simulate this by starting with a string in that format. Write code that will extract the x, y, and z coordinates as numbers. To verify that they are numbers, add them together. Hint: use the **index** method to find the locations of the characters.

```python
# 13
oneline = 'x12.34y3.5z4.21'
oneline = oneline[1:]  # skip over 'x'
wherey = oneline.index('y')
x = float(oneline[:wherey])
restline = oneline[wherey+1:]
wherez = restline.index('z')
y = float(restline[:wherez])
z = float(restline[wherez+1:])
print(x + y + z)
```
