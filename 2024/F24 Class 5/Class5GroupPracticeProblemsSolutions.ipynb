In the last class we covered ways to execute code, comments, user input, formatting output, and debugging.  Today we will cover: selection, specifically, if statements and if-else statements.

Sometimes, we want to be able to decide whether or not to run a line or a block of code.  Statements that accomplish this are called **selection** statements.  We are going to focus on **if** and **if-else** statements here, but python includes additional selection statements, too.

The if statement chooses whether statement(s) are executed or not.  The syntax of an if statement is as follows:

    if condition:

        action

    rest of code

"condition" is an expression that must evaluate to True or False.

"action" must be intented.  It is only executed if the condition is true.  Otherwise, it is skipped.

Here's an example.  Try it with several different ages.

```python
#1
age_s=input('What is your age')
age=int(age_s)
if age < 30:
    print('Never trust anyone over 30')
print('And the code moves on')
```

We can also choose between conditions using if-else statements.  The general form is

    if condition:
       ifaction
    else:
       elseaction
    # rest of code

This code will choose between two actions, "ifaction" and "elseaction" depending on whether or not the condition is true.

Here's an example.  Try it out.

```python
#7
name=input('Please enter your name: ')
if name.lower() == 'arthur':
    print('You are the king of the Britons!')
else:
    print('You are a brave knight on a quest for the Holy Grail!')
```

It is also possible to nest if/else statements so you can check multiple conditions.  

For example, let's say we wanted to check to see if a given year was a leap year.  Leap years must be divisible by 4, unless they are century years, then they must also be divisible by 400.  The code to  check a user input year would look like this.  Try it out.

```python
#9
year = int(input("Enter a year: "))
if year % 4 == 0:
  if year % 100 == 0:
    if year % 400 == 0:
      print('Leap year!')
    else:
      print('Not a leap year.')
  else:
    print('Leap year!')
else:
  print('Not a leap year.')
```

This type of statement can be simplified using the elif keyword.

```python
#10
# Prompt the user for the number of limbs lost
limbs_lost = int(input("Enter the number of limbs the Black Knight has lost: "))

# Determine the message to print
if limbs_lost == 0:
    message = "The Black Knight says: 'None shall pass!'"
elif limbs_lost == 1:
    message = "The Black Knight says: 'It's just a flesh wound!'"
elif limbs_lost == 2:
    message = "The Black Knight says: 'I've had worse!'"
elif limbs_lost == 3:
    message = "The Black Knight says: 'Come on, you pansy!'"
else:
    message = "The Black Knight says: 'Alright, we'll call it a draw.'"

# Print the message
print(message)
```

Try witing an age group classifier using the elif keyword.  Prompt the user for an age, and then classify the person based on age ranges: 0-12 = child, 13-19 = teen, 20-64 = adult, 65 and over = senior.

```python
#11
# Prompt the user for the age
age = int(input("Enter the person's age: "))

# Determine the message to print
if age < 13:
    message = "The person is a child"
elif 13 <= age <= 19:
    message = "The person is a teenager"
elif 19 < age < 65:
    message = "The person is an adult"
else:
    message = "The person is a senior"

# Print the message
print(message)
```
