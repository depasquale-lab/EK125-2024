Write a function that will receive a string that contains several sentences, separated by \n characters. The function should capitalize the first word in every sentence, and return a list of the capitalized sentences.

```python
# Problem 1

def capitalize_sentences(text):
    """ Capitalizes the first word in every sentence in the input text."""
    sentencelist = text.splitlines()
    newlist = [sentence.capitalize() for sentence in sentencelist]
    return newlist

sents = "hi.\nthis will be a test.\ni hope it works."
print(capitalize_sentences(sents))
```

Create two string variables that consist of a character, a space, and a number, in the form 'R 14.3'. Write a function that will receive these, and return the sum of the numbers in them.

```python
# Problem 2

def addem(str1, str2):
    """ Extract the numbers from strings of form 'R 14.3' and return their sum."""
    num1 = float(str1.split()[1])
    num2 = float(str2.split()[1])
    return num1 + num2

addem('R 14.3', 'Q 22.2')
```

Write a script that will generate a random integer in the range from 1 to 3. It will then call a function to prompt the user for that many strings, and return a list of all of the strings. It will then call a function that will capitalize all of the words in the first string, and only the first word in the rest, and return the result.

```python
# Problem 3

def read_strings(n):
    """ Prompts the user for n strings and returns a list of them."""
    stringlist = []
    for i in range(n):
        stringlist.append(input("Enter a string: "))
    return stringlist

def cap_some(strlist):
    """ Capitalizes all words in the first string,
    and only the first word in the rest."""
    strlist[0] = strlist[0].title()
    for i in range(1, len(strlist)):
        strlist[i] = strlist[i].capitalize()
    return strlist

from random import randint
n = randint(1, 3)

stringlist = read_strings(n)
print(cap_some(stringlist))
```

Write a function that will receive a string, and will print how many times every letter of the alphabet is in the string. No, you are NOT going to have 26 different counters!  (If you ever find yourself doing something like that, immediately think to yourself "there must be a better way"). Create a count list instead, which will be initialized to all 0's.

```python
# Problem 4

def countalpha(text):
    """ Counts and prints the number of times each letter
    of the alphabet is in the text."""
    from string import ascii_lowercase
    countlist = [0] * 26
    for char in text:
        if char.lower() in ascii_lowercase:
            countlist[ascii_lowercase.index(char.lower())] += 1
    for i in range(26):
        print(f"{ascii_lowercase[i]}: {countlist[i]}")

charstr = "The quick brown fox jumps over the lazy dog."
countalpha(charstr)
```

Write a function that receives a string and compresses it using the counts of repeated characters. For example, the string ‘aabcccccaaa’ would become ‘a2b1c5a3’.

```python
def compress_string(s):
    # Initialize variables
    compressed = []
    count = 1

    # Iterate through the string
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed.append(s[i - 1] + str(count))
            count = 1

    # Add the last character and its count
    compressed.append(s[-1] + str(count))

    # Join the list into a string and return
    return ''.join(compressed)

# Example usage
input_string = 'aabcccccaaa'
compressed_string = compress_string(input_string)
print(f"Compressed string: {compressed_string}")
```
