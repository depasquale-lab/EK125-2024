**Problem 4**:  What is the difference between the following statements?

>>> import math

>>> from math import *

The first line gives access to the module's functions and definitions.  They will still need to be called using the dot operator.  This keeps them distinct from any user-defined variables or functions.

The second line imports each of the functions and definitions directly so they can be referenced by name.  The danger here is that if you don't know all of the names of the items in the module, you may run the risk of a conflict with some of your own variable names.
