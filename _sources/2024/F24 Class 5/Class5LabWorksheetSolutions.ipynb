**Problem 1**:  Write a program that takes the total purchase amount and membership status (e.g., “Gold”, “Silver”, “None”) as input. Use nested if statements to calculate and print the final price after applying the appropriate discount based on the membership status.  Assume a Gold membership comes with a 15% discount and and a 20% discount on purchases of 100 dollars or more. A Silver membership has an 5% discount  and a 10% discount on purchases of 100 dollars or more.  Print out the total amount, the discount, and the final price after discount formatted in dollars and cents ($dd.cc).  Comment your program accordingly.

```python

# Get user input for total purchase amount and membership status
total_amount = float(input("Enter the total purchase amount: "))
membership_status = input("Enter membership status (Gold/Silver/None): ")

# Initialize discount variable
discount = 0.0

# Determine discount based on membership status
if membership_status == "Gold":
    if total_amount >= 100:
        discount = 0.20  # 20% discount for Gold members on purchases of $100 or more
    else:
        discount = 0.15  # 15% discount for Gold members on purchases below $100
elif membership_status == "Silver":
    if total_amount >= 100:
        discount = 0.10  # 10% discount for Silver members on purchases of $100 or more
    else:
        discount = 0.05  # 5% discount for Silver members on purchases below $100
elif membership_status == "None":
    if total_amount >= 100:
        discount = 0.05  # 5% discount for non-members on purchases of $100 or more
    else:
        discount = 0.0   # No discount for non-members on purchases below $100
else:
    print("Invalid membership status.")

# Calculate final price after discount
final_price = total_amount * (1 - discount)

# Print the final price
print(f"Total amount: ${total_amount:.2f}")
print(f"Discount: {discount * 100:.0f}%")
print(f"Final price after discount: ${final_price:.2f}")
```

** Problem 2 ** Write a Python program that takes a string input from the user and performs the following operations based on the content of the string:

If the string contains only alphabetic characters, print “Alphabetic string”.

If the string contains only digits, print “Numeric string”.

If the string contains both alphabetic characters and digits, print “Alphanumeric string”.

If the string contains any special characters (neither alphabetic nor digits), print “String with special characters”.

Hint: Remember your string methods...

```python
# Get input from the user
user_input = input("Enter a string: ")

# Check the content of the string and perform operations
if user_input.isalpha():
    print("Alphabetic string")
elif user_input.isdigit():
    print("Numeric string")
elif user_input.isalnum():
    print("Alphanumeric string")
else:
    print("String with special characters")
```

**Problem 3**:  Write a program that allows the user to enter a numerical grade and have the letter grade returned.  Whole letter grades are sufficient.  You may determine the score ranges.  Please comment your code appropriately.

```python
# Get the score from the user
score = float(input("Enter the student's score (0-100): "))

# Determine the grade based on the score
if score >= 90 and score <= 100:
  grade = 'A'
elif score >= 80 and score < 90:
  grade = 'B'
elif score >= 70 and score < 80:
  grade = 'C'
elif score >= 60 and score < 70:
  grade = 'D'
elif score >= 0 and score < 60:
  grade = 'F'
else:
  grade = 'Invalid score'

# Print the grade
print(f"The grade for a score of {score} is: {grade}")

```
