A great use of for loops is to loop through a sequence, performing the same action on each element in the sequence.  Create a list of float numbers. Write a for loop that will loop through the list and print the floor of each number. Recall that floor is in the math module.

```python
# 9
from math import floor
myflist = [3.3, 4.5, 11.11, 2.9, -5.2]
for i in myflist:
    print('The floor of ',i,'is',floor(i))
```

An application of for loops is to calculate running sums. The following code calculates and prints the sum of the integers from 0 to 3. Execute it.

```python
# 12
runsum = 0
for i in range(4):
    runsum = runsum + i
print('The sum is', runsum)
```

Now, modify the code so that instead of printing the overall sum, it prints each running sum as it is calculated.

```python
# 13
runsum = 0
for i in range(4):
    runsum = runsum + i
    print('The sum so far is', runsum)
```

 A running product is the product of numbers. A running sum is initialized to 0, so the first time something is added to it, it does not change. A running product, on the other hand, is initialized to 1. The following should calculate a running product of the integers from 1 to 5. Execute it. What is wrong?

```python
# 14
runprod = 1
for i in range(6):
    runprod = runprod * i
    print('The product so far is', runprod)
```

Fix the code!

```python
# 15
runprod = 1
for i in range(5):
    runprod = runprod * (i+1)
    print('The product so far is', runprod)
```
