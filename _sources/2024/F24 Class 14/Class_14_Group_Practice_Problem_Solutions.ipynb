Write a sign function. The sign of a positive number is 1, the sign of 0 is 0, and the sign of a negative number is -1. Be sure to test all three cases.

```python
# 2

def sign(num):
    """ This function returns the sign of a number"""
    if num > 0:
        return 1
    elif num == 0:
        return 0
    else:
        return -1


print(sign(10))
print(sign(0))
print(sign(-10))
help(sign)
```

```python
# 3

def get_float(min_val, max_val):
    """ This function prompts the user for a float in a range"""
    if min_val > max_val:
        temp = min_val
        min_val = max_val
        max_val = temp
    urnum = float(input(f"Enter a number between {min_val} and {max_val}: "))
    while urnum < min_val or urnum > max_val:
        print("Invalid input")
        urnum = float(input(f"Enter a number between {min_val} and {max_val}: "))
    return urnum


help(get_float)
print("Your number was", get_float(1, 10))
```

Write a function that will receive two arguments: the length and width of a rectangle. The function will return the area and perimeter of the rectangle in a tuple. Test calling the function and storing the area and perimeter that are returned in separate variables.

```python
# 4

def rect_area_peri(length, width):
    """ This function returns the area and perimeter of a rectangle"""
    area = length * width
    perimeter = 2 * (length + width)
    return (area, perimeter)


print(rect_area_peri(2, 3))
help(rect_area_peri)
```

Write a function that will print a box of *'s. Two arguments will be passed to the function: the number of rows to print, and the number of columns (the number of *'s in each row).

```python
# 5
def print_box(rows, cols):
    """ This function prints a box of *'s"""
    for i in range(rows):
        for j in range(cols):
            print("*", end="")
        print()
    print()


print_box(3, 5)
help(print_box)
```

Write a cumulative product function. The function will receive a list of numbers, and will create a new list which is a list of the running products of the numbers in the parameter list. For example, if the list that is passed to the function is [4, 2, 5], the returned list would be [4, 8, 40].

```python
# 8

def cum_prod(num_list):
    """ This function returns a list of the running products of the numbers in the parameter list"""
    prod_list = []
    prod = 1
    for num in num_list:
        prod = prod * num
        prod_list.append(prod)
    return prod_list


print(cum_prod([4, 2, 5]))
print()
help(cum_prod)
```

Since strings are immutable, we cannot use the reverse method with strings. Write a function string_rev that will receive a string parameter and will reverse the characters in the string and return the result.

```python
# 9

def string_rev(string):
    """ This function reverses the characters in a string and returns the result"""
    rev_string = ""
    for char in string:
        rev_string = char + rev_string
    return rev_string

newstr = string_rev("hello")
print(newstr)
print()
help(string_rev)
```

Write a function that will shuffle the elements in a tuple and return the result.

```python
# 10

def shuffle_tuple(tup):
    """ This function shuffles the elements in a tuple and returns the result"""
    import random
    new_tup = list(tup)
    random.shuffle(new_tup)
    return tuple(new_tup)

print(shuffle_tuple((1, 2, 3, 4, 5)))
print()
help(shuffle_tuple)
```

Write an "interpolate" function, which takes two tuples, *(x0,y0)* and *(x1,y1)*, and an input *x_i*. It will then return the "interpolated" y value, that is to say the one which would correspond to the x_i as if it were on a line. In other words, if (x0,y0) = (0,0), and if (x1,y1) = (1,0.5), then passing x_i=0.5 would result in returning y_i=0.25. And passing x_i=0.9 would return y_i = 0.45.

It should return an error for any x value outside the range \[*x0,x1*\].

```python
#12
def interpolate(X0, X1, x_i):
  """ This function interpolates between two points"""
  # Do some bounds checking
  if x_i < X0[0]:
    print ("Error, ", x_i, " is less than ", X0[0])
    return None
  elif x_i > X1[0]:
    print ("Error, ", x_i, " is greater than ", X1[0])
    return None

  # Calculate the slope
  slope = (X1[1]-X0[1])/(X1[0]-X0[0])
  y_i = X0[1] + slope*(x_i - X0[0])

  return y_i

interpolate((10,1),(20,2), 19)
# help(interpolate)
```
