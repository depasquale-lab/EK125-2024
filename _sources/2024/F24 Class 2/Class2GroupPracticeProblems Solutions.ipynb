The order of precedence for operations covered so far:Parentheses, Exponentiation, Negation, Multiplication/Division, Addition/Subtraction, Relational, Not, And, Or.

Also like in algebra, operations with the same precedence are evaluated left to right.

Try these examples:

```python
(3+1**2)*(2/4+3)
```

```python
(3+1)**2*2/4+3
```

```python
new_val = (5 < 6) and not (6 > 7)
new_val = new_val + 7 * 3
new_val
```

Sometimes we want to be able to store more than value in a specific order, and we can do that in a data structure called a **sequence**.  We'll revisit these later in the course, but for now, it will be useful to think about **strings** and **lists** as sequences of characters or elements that are put in order and can be accessed using a variable name and an integer index.

A string is a sequence of characters, and can be stored in either single quotes or double quotes.  The **print** function below can print combinations of strings.

```python
string1 = 'What is the airspeed'
string2 = "of an unladen swallow?"
print(string1, string2)
string1
```

Each character in the string is labeled, and can be accessed, with an integer starting at 0.  These integers are known as indices.

We can index into the string by putting the character's index into square brackets.  

```python
print(string1[0], string1[10], string2[3], string2[19])
```

The **len** function returns the length of a string.

How many indices does string1 above have?  

```python
len(string1)
```

What is the largest index of string2?  

```python
len(string2)
```

```python
Largest index: 22
```

There are some operations we can perform on strings.  The **in** operator will determine if one string is contained within another string.  Strings can be concatenated (or joined) using the '**+**' operator.

First, let's put our two short strings together to create a full sentence.

```python
sentence=string1 + " " + string2
sentence
```

Now let's check to see if the word "wall" is contained in our concatenated string.

```python
string3='wall'
string3 in sentence
```

You might think that because you can index each character in the string separately, you should be able to modify each character separately.  Let's see if we can change string3 from "wall" to "ball" by changing the first character.

```python
string3[0] ='b'
```

Hmmm.  No dice.  Strings are immutable.  They cannot be changed once they are created.  If you try to change an element in a string, you'll get an error.

You can also compare strings with relational operators.

```python
'swallow' > 'plover'
```

```python
's' > 't'
```

**Lists** are also sequences of values.  These values can be of different types.  Lists are created by putting values in square brackets separated by commas. Run the code below.

```python
numlist = [1, 2, 5, 3, 3]
print(numlist, type(numlist))
```

The values in this list are all numbers, but the type of the list is still **list**.

```python
GrenadeofAntioch = ["right", 1, 2, 5, 'three, sir', 3]
print(GrenadeofAntioch, type(GrenadeofAntioch))
```

Lists can still be indexed using integers.  Unlike strings, they are mutable, so they can be modified.  Let's try changing the last element of our list.

```python
GrenadeofAntioch [len(GrenadeofAntioch)-1] = "boom"
print(GrenadeofAntioch)
```

Notice that we've used the **len** function in the code above.  Len will return the number of elements in a list.  It will also give you the number of elements contained in a single element of your list. For example:

```python
len(GrenadeofAntioch[4])
```

Try using the **in** operator like we did with strings to see if the string 'boom' is included in the list GrenadeofAntioch.

```python
"boom" in GrenadeofAntioch
```

Finally, each expression in our list has its own type depending on the form of the expression.  But what if we want to convert that expression to another type?  We can using explicit **type casting**.  To type cast, simply use the type name as a function.

```python
x = '4.3'
print(x, type(x))
x = float(x)
print(x, type(x))
x = int(x)
print(x, type(x))
```
