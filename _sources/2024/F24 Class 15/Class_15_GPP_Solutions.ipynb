### OK, now let's start looking at some of the ramifications of this.



Let's imagine that we have a list which we want to reverse, does this create a new object in memory or does it use the prior one? Knowing what you know about assignments and copying, take a second and make a prediction about whether a new copy **is made** or a new copy **is not made**.

Execute the code below and test your hypothesis.

````python
#
print(id(myList), myList)
myList.reverse()
print(id(myList), myList)
````

There's a concept called a *deep copy* and another one called a *shallow copy*. A *deep copy* is when everything in the object is copied over to the new object. The *shallow copy* is when only the reference to the object is copied. Let's look at how this works:

#### Deep copy
Note after the assignment how changing `a` doesn't affect `b`. They're totally separate objects.

````python
#
a = 5
b = a
a = 3
print(a,b)
````

#### Shallow copy

Note how the two lists are not at all dependent!

````python
# Set the myList
myList = ['1', '2', '3', '4']

# Create a new variable, `myNewList`, and assign `myList` to it.
myNewList = myList
myNewList[0] = '0'

# Print the old list
print("myList: ", myList)

# Check that these two variables, `myList` and `myNewList` are indeed exactly the same!
print(myNewList is myList)
````

Oh, okay, that's a bit of a kick in the pants. So sometimes you *think* you've made a deep copy, but that's not always the case. So when objects are mutable, and you change just a single index in the object, you do **not** create a new object.

Honestly, it's a bit hard to explain, right now we just need you to understand that it happens.


BTW, what about if you don't want this to happen? What if you wanted to make a list, and you wanted to `reverse()` a copy of the list without modifying the original?

````python
# Set the stage:
print("myList: ", myList)
print("myNewList: ", myNewList)

# Act:
myList.reverse()

# Check results
print("myList: ", myList)
print("myNewList: ", myNewList)

# Do a DEEP COPY
myNewList = myList.copy()

# Set the stage:
print("myList: ", myList)
print("myNewList: ", myNewList)

# Act:
myList.reverse()

# Check results
print("myList: ", myList)
print("myNewList: ", myNewList)
````

### Tuples

Run the below and notice that once again we have a **shallow** copy, but unlike with the list we cannot change the elements. So it doesn't really matter if we have a shallow or deep copy, as the result is exactly identical.
