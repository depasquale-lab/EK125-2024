Write a function which converts from degrees to radians, taking as input `armAngle_deg` and returning `armAngle_rad`.

```python
# Problem 4
import math

def degrees_to_radians(armAngle_deg):
    # Use math.pi to get the value of pi
    armAngle_rad = armAngle_deg * (math.pi / 180)
    return armAngle_rad

# Example usage
angle_in_degrees = 45
angle_in_radians = degrees_to_radians(angle_in_degrees)
print(f"{angle_in_degrees} degrees is equal to {angle_in_radians} radians.")
```

Write a function that finds the median value in any list.

```python
# Problem 5
def find_median(values):
    # Sort the list first
    sorted_values = sorted(values)
    length = len(sorted_values)

    # If the list has an odd number of elements, return the middle one
    if length % 2 == 1:
        median = sorted_values[length // 2]
    else:
        # If the list has an even number of elements, return the average of the two middle ones
        median = (sorted_values[length // 2 - 1] + sorted_values[length // 2]) / 2

    return median

# Example usage
example_list = [7, 2, 9, 4, 3]
median_value = find_median(example_list)
print(f"The median is: {median_value}")
```

Implement a simple "Caesar cipher", i.e. one where you shift characters by a certain number of places. So if the number is there, and the character is `a`, then the output is `d`. This wraps at the end, so the character `z` becomes `c`.

```python
# Problem 6
def caesar_cipher(text, shift):
    result = ""

    # Loop through each character in the input text
    for char in text:
        # Check if the character is a lowercase letter
        if char.islower():
            # Shift the character and wrap around using modulo
            result += chr((ord(char) - ord('a') + shift) % 26 + ord('a'))
        # Check if the character is an uppercase letter
        elif char.isupper():
            # Shift the character and wrap around using modulo
            result += chr((ord(char) - ord('A') + shift) % 26 + ord('A'))
        else:
            # If it's not a letter, keep it unchanged
            result += char

    return result

# Example usage
text = "abc XYZ z"
shift = 3
encrypted_text = caesar_cipher(text, shift)
print(f"Encrypted text: {encrypted_text}")
```

Write a function which when provided with a number, or list of numbers, between 0-9, return the number or list of numbers in words. In other words, if given `0`, the output is `zero`. If given `[0,0,1]` the output is `['zero', 'zero', 'one']`. Note that the input is guaranteed to be within the range of 0-9.

```python
# Problem 7
def number_to_word(n):
    if n == 0:
        return "zero"
    elif n == 1:
        return "one"
    elif n == 2:
        return "two"
    elif n == 3:
        return "three"
    elif n == 4:
        return "four"
    elif n == 5:
        return "five"
    elif n == 6:
        return "six"
    elif n == 7:
        return "seven"
    elif n == 8:
        return "eight"
    elif n == 9:
        return "nine"

def numbers_to_words(input_value):
    if isinstance(input_value, list):
        # If input_value is a list, convert each number in the list
        return [number_to_word(n) for n in input_value]
    else:
        # If input_value is a single number, convert it
        return number_to_word(input_value)

# Example usage
print(numbers_to_words(5))         # Output: 'five'
print(numbers_to_words([0, 0, 1])) # Output: ['zero', 'zero', 'one']
```

Write a function to reverse a given string without using built-in functions.

```python
# Problem 8
def reverse_string(s):
    # Initialize an empty string to hold the reversed result
    reversed_str = ""

    # Loop through the original string from the last character to the first
    for i in range(len(s) - 1, -1, -1):
        reversed_str += s[i]

    return reversed_str

# Example usage
input_string = "hello"
reversed_result = reverse_string(input_string)
print(f"The reversed string is: {reversed_result}")
```
