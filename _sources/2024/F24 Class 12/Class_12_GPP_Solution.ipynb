The enumerate function returns the index and item from a sequence. Execute the following:

```python
# 12
list(enumerate([3, 2, 11]))
```

This actually created a list of tuples (which we'll cover shortly!). In each, you have the index and then the item from the list. Let's use this in a for loop. Execute the following:

```python
# 13
for index, item in enumerate([3, 2, 11]):
    print(index, item)
```

Use enumerate and a for loop to print the indices and individual characters from a string.

A nested list is created for you. Use nested for loops to print each item and indices in the format "Item 0 in list 0 is 5", then "Item 1 in list 0 is 11", etc.

```python
# 15
nestlist = [[5, 11], [4, 2, 33], list(range(2,6))]

for i, onelist in enumerate(nestlist):
    for j, oneval in enumerate(onelist):
        print(f'Item {j} in list {i} is {oneval}')
```

Create a list of strings. Then, for each string, print the index of the string (naturally, e.g. 1,2,3.. instead of 0,1,2..), the number of characters in the string, and the individual characters. For example, if the first string is "Hello", the first line of output would be "Word 1 "Hello" has 5 characters: H e l l o"

```python
# 16
strlist = ["Hello", "Hi", "Ciao"]
for i, s in enumerate(strlist):
    print(f'Word {i+1} "{s}" has {len(s)} characters:', end=' ')
    for c in s:
        print(c, end=' ')
    print()
```
