Problem 4: One use of the enumerate function is to loop through two sequences that have the same length simultaneously, performing an operation on the items with the same indices. For example, the following adds corresponding elements in two lists, creating a third list with the results.

```python
# 4
a = range(1,6)
b = [2, 11, -5, 4, 33]
c = []
for i, item in enumerate(a):
    c.append(item + b[i])
c
```

Problem 5: It is frequently useful to calculate the differences between consecutive elements in a list of numbers. For example, if the list is [5, 11, 2, 33, -4], the differences would be [11-5, 2-11, 33-2, -4-33], or [6, -9, 31, -37]. For a list variable that has a length of n, the resulting differences list would have a length of n-1. Using a list variable numlist, start by creating two list variables: one stores all but the first element, and the other stores all but the last element. Then, using the method above, use a for loop and enumerate to print the differences.

```python
# 5
numlist = [5, 11, 2, 33, -4]
list1 = numlist[1:]
list2 = numlist[:len(numlist)-1]
for i, item in enumerate(list1):
    print(item - list2[i], end = ' ')
print()
```

Problem 6: A list variable gfcc stores for several employees of the Green Fuel Cells Corporation their hours worked one week followed for each by the hourly pay rate.  For example, if the variable stores [33,  30.5,  40,  28,  20,  27.5] that means the first employee worked 33 hours at 30.50 per hour, the second worked 40 hours at 18 an hour, and so on.  Write code that will separate this into two lists, one that stores the hours worked and another that stores the hourly rates.  Then, calculate the total pay for every employee.  


```python
# 6
gfcc = [33, 30.5, 40, 28, 20, 27.5]
hours = gfcc[::2]
rates = gfcc[1::2]
for i, hr in enumerate(hours):
    print(f'Pay is ${hr*rates[i]:.2f}.')
```

Problem 7: Create a nested tuple of numbers. Find the overall maximum.


```python
# 7
numstup = ((2, 6, -1), (33, 5, 14), (-99, 52))
maxsofar = numstup[0][0]
for onetup in numstup:
    if max(onetup) > maxsofar:
        maxsofar = max(onetup)
print('The overall maximum is', maxsofar)

```
