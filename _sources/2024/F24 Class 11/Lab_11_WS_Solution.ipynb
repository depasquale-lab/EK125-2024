**Problem 2**: Write a program that removes the duplicate items from a list.  The program should return a list with the duplicate items removed but maintain the original order of the list.  The program should also print a list of the items that were removed.

```python
# 2
# enter a list of items of your choosing for usrlist
usrlist=[]
unique_items = []
duplicates = []
for item in usrlist:
  if item not in unique_items:
    unique_items.append(item)
  else:
    duplicates.append(item)
print('The unique items in the list are:', unique_items)
if len(duplicates) == 0:
  print('There were no duplicate items.')
else:
  print('The duplicate items in the list are:', duplicates)
```

**Problem 3**:  Write a program that prompts the user for two strings.  The program should then check to see if the two strings are anagrams of each other.  The program should also check to see if either of the two strings is a palindrome.  Finally, print the result of your checks.

```python
# 3

str1=input('Please enter a string: ')
str2=input('Please enter a second string: ')

is_anagram = (sorted(str1) == sorted(str2))

s1_is_palindrome = (str1 == str1[::-1])
s2_is_palindrome = (str2 == str2[::-1])

if is_anagram:
  print('The two strings are anagrams of each other.')
else:
  print('The two strings are not anagrams of each other')
if s1_is_palindrome:
  print('String 1 is a palindrome.')
if s2_is_palindrome:
  print('String 2 is a palindrome.')
```

**Problem 4**: Write a program that prompts the user to enter a string or list and then a character or element in the list.  The program should count and print the number of instances of that element and if that count is not zero, should print  all the locations (indices) of that character/element in the sequence.

```python
# 4
sequence=input('Please enter a sequence: ')
element=input('Please enter an element to find: ')
count=sequence.count(element)


if count == 1:
  print('The element appears 1 time.')
else:
  print('The element appears', count, 'times.')


if count > 0:
  indices=[]
  for i in range(len(sequence)):
    if sequence[i] == element:
      indices.append(i)
  if len(indices) == 0:
    print('The element does not appear in the sequence.')
  else:
    print('The element appears at the following indices:', indices)
```

**Problem 5**:  When checking the help documentation on the sorted method, one finds that a custom key function can be supplied to customize the sort order.  If you want to use a built in function, you can simply use the function name as the key (i.e. key=sum)  You can also define a custom function with the syntax: key=lambda x: + your definition here (i.e key=lambda x: x.lower())  Try defining some of your own keys to sort lists by different criteria.  For example, sort a list of strings by length, or by last character.

```python
# 5
strings = ["banana", "apple", "cherry", "date"]
# sorting by last character
new_string=sorted(strings, key=lambda x: x[-1])
print(new_string)

#sorting by length
new_string=sorted(strings, key=len)
print(new_string)
```

Problem 6: Write a Python program that defines a list of numbers. The program should prompt the user to enter a number and a position, then insert the number at the specified position in the list. Finally, print the updated list.

```python
# 6
# Define the list of numbers
numbers = [10, 20, 30, 40, 50]

# Prompt the user to enter a number and a position
num = int(input("Enter a number to insert: "))
position = int(input("Enter the position to insert the number at: "))

# Check if the position is valid
if 0 <= position <= len(numbers):
    numbers.insert(position, num)
    print("Updated list:", numbers)
else:
    print("Invalid position.")
```

Problem 7: Write a Python program that defines a list of numbers. The program should prompt the user to enter a position and remove the element at that position using the `pop` method. If the position is invalid, display an appropriate message. Finally, print the updated list.

```python
# 7
# Define the list of numbers
numbers = [11, 22, 33, 44, 55]

# Prompt the user to enter a position
position = int(input("Enter the position to remove an element from: "))

# Check if the position is valid and pop the element
if 0 <= position < len(numbers):
    removed_element = numbers.pop(position)
    print(f"Removed element: {removed_element}")
    print("Updated list:", numbers)
else:
    print("Invalid position.")
```
