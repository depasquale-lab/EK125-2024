**Problem 1**:  Write a short program to ask the user for two integers.  Calculate and display the product of those integers in a complete sentence.  Comment your code appropriately.

```python
#Prompt the user for two integers
int1 = int(input("Please enter an integer: "))
int2 = int(input("Please enter a second integer: "))

product = int1 * int2
print(f'The product of the two integers you entered is {product}.')
```

**Problem 2**: Write a short program to ask the user for two floats with at least two decimal places each.  Display the product of those two numbers with exactly three decimal places in a complete sentence.  Comment your code appropriately.

```python
#Prompt the user for two floats with two decimal places
float1 = float(input("Please enter a floating point number with at least two decimal places: "))
float2 = float(input("Please enter a second floating point number with at least two decimal places: "))

product = float1 * float2
print(f'The product of the two integers you entered is {product:.3f}.')
```

**Problem 3**:  Write a program that calculates the simple interest on a principal amount. Prompt the user for the principal amount, the yearly interest rate and the amount of time in years.  Display the output in a complete sentence.  Comment your code appropriately.

```python
# Prompt the user for the principal amount, rate of interest, and time period
principal = float(input("Enter the principal amount: "))
rate = float(input("Enter the rate of interest: "))
time = float(input("Enter the time period in years: "))

# Calculate the simple interest
simple_interest = (principal * rate * time) / 100

# Print the result
print(f"The simple interest is {simple_interest:.2f}.")
```

**Problem 4**:  Sir Lancelot would like a program to determine the area of his round table and determine the number of knights who can comfortably sit around it.  This program should prompt the user for the diameter of the table (it's the easiest thing to measure) and should print the area with a reasonable number of decimal places in a full sentence and the number of knights in another full sentence.  The average knight will need 2.5 feet of space to sit comfortably (Armor, you know).  Comment this program appropriately.

```python
# We'll need pi to calculate area and circumference of a circle
import math

#Prompt for the table diameter
diameter=float(input('Sir, how large is your table?  Diameter in feet, please: '))

#Area = pi*(D/2)^2
area = math.pi*(diameter/2)**2
circumference = math.pi*diameter
knights=math.floor(circumference/2.5)
circumference
#Print the answers in full sentences
print(f'The area of your table is {area:.2f} square feet.')
print(f'{knights} knights can be comfortably seated around your table.')
```
