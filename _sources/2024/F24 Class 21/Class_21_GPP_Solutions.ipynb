The trace of a diagonal matrix is the sum of the numbers on the diagonal. Write a function that will receive a diagonal matrix and return its trace. Create a 4 by 4 diagonal matrix, and pass this to your function.

```python
# 15

def mytrace(diagmat):
    """ Returns the trace of a diagonal matrix"""
    x = np.diag(diagmat)
    return np.sum(x)

diag44 = array([[3, 0, 0, 0], [0, 7, 0, 0], [0, 0, 6, 0], [0, 0, 0, 11]])
print(diag44)
print(mytrace(diag44))
```

Write a function that will receive as input a 2D array. It will replace the first row of the array with the product of the orignal first row and 2, and return the result. Make sure that the function does not alter the array that is passed to it.

```python
# 18

def row0times2(inmat):
    """Replaces the first row of the matrix with row * 2"""
    mat = np.copy(inmat)
    mat[0,:] = mat[0,:] * 2
    return mat

mymat = array([[1, 3], [2, 1]])
print(mymat)
print(row0times2(mymat))
print(mymat)
```
