Now, with a list of positive integers, write a nested for loop to loop through the list. For every element, print that many exclamation points on one row, with no spaces in between. Write general code. If the list of integers is as specified below, the first line would have 3 exclamation points, the second line would have 7, etc. Do this two ways: by having the outer loop loop through the elements in the list, and by having the outer loop loop through the indices.

```python
# 5

intlist = [3, 7, 2, 1, 5]

for i in intlist:
    for j in range(i):
        print('!', end = '')
    print()

print('\n\n')

for i in range(len(intlist)):
    for j in range(intlist[i]):
        print('!', end = '')
    print()
```

Write a script that will prompt the user for positive numbers. As long as the user is entering positive numbers, Print 'Ni!' that many times on a single line for each. For example, if the user enters 3, 'Ni!Ni!Ni!' would be printed.

```python
# 7

posnum = int(input('Enter a positive number: '))

while posnum > 0:
    for i in range(posnum):
        print('Ni!', end = '')
    print()
    posnum = int(input('Enter a positive number: '))
```

Generate a random integer, n,  in the range from 2 to 5. Then, n times, prompt the user for a positive number. Each time, error-check to make sure that the user enters a positive number.

You're an overworked and underappreciated server at [The Green Midget Cafe](https://www.youtube.com/watch?v=anwy2MPT5RE).  Trying to repeat the menu to customers over a bunch of chanting Vikings is straining your voice, so you want to automate generating the menu.  

1. Eggs and bacon
2. Eggs, sausage, and bacon
3. Eggs and spam
4. Eggs, bacon, and spam
5. Eggs, bacon, sausage, and spam
6. Spam, bacon, sausage, and spam
7. Spam, eggs, spam, spam, bacon, and spam
8. Spam, spam, spam, eggs, and spam
9. Spam, spam, spam, spam, spam, spam, spam, baked beans, spam, spam, spam, and spam
10. Spam, sausage, spam, spam, spam, bacon, spam, tomato, and spam


Write a script that will generate each menu item from the list provided and print them as a numbered list so your customers can simply order by number.  Also, you're British, so that Oxford comma is important!


```python
#10
# Use this list of items to create your nested loops
ingredients = ['eggs', 'bacon', 'sausage', 'spam', 'baked beans', 'tomato']
indices = [[0,1], [0,2,1], [0,3], [0,1,3], [0,1,2,3], [3,1,2,3], [3,0,3,3,1,3], [3,3,3,0,3], [3,3,3,3,3,3,3,4,3,3,3,3], [3,2,3,3,3,1,3,5,3]]
menu = [''] * len(indices)  # Initialize menu with empty strings

for i in range(len(indices)):
    if len(indices[i]) == 2:
        menu[i] = f'{ingredients[indices[i][0]]} and {ingredients[indices[i][1]]}'
    else:
        for j in range(len(indices[i]) - 1):
            menu[i] += ingredients[indices[i][j]] + ', '
        menu[i] += 'and ' + ingredients[indices[i][-1]]
    print(f"{i + 1}. {menu[i]}")

print(f"Total menu items: {len(menu)}")
```
